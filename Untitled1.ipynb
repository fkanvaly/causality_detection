```python
import networkx as nx
import matplotlib.pyplot as plt
```

```python
graph = nx.complete_graph(4)
G = nx.DiGraph(list(graph.edges()))
DiG = nx.DiGraph(list(G.edges()) + list(G.reverse().edges()))
DiG.remove_edge(2,3)
DiG.remove_edge(3,2)
for path in nx.all_simple_paths(DiG, source=0, target=3):
    print(path)
```

```python
nx.draw(DiG, with_labels = True)
plt.show()
```
